We can draw samples from the posterior using sample, ======================================================but we first need to 'burn in' the MCMC until it converges to a stationary distributionand thin the chain to overcome auto-correlations in the samples

```python
# Run an MCMC chain of length 10,000 with 1000 burn-in and thinning by 10# using Metropolis-Hastings algorithmM.sample(iter=10000, burn=1000, thin=2)# isample does interactive MCMC fitting, sample does non-interactive fitting
```
